```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
        'A': ['foo','foo','foo', 'foo', 'foo', 'bar', 'bar','bar','bar'], 
        'B': ['one','one','two', 'two', 'one','one','two','two', 'one' ], 
        'C': ['small', 'large', 'large', 'small', 'small', 'large', 'small', 'small', 'large'], 
        'D': [1,2,2,3,3,4,5,6,7]
    })
df
```

```python
table = pd.pivot_table(df, values='D', index=['A','B'], columns='C', aggfunc=np.sum)
table
```
